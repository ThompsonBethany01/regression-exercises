## What probability distribution would you use to model the scenario outlined above?
- Calculate all the requested probabilities. Use all the possible combinations of subject count and chance that a subject will stay in the study.

```python
from scipy import stats
```

```python
# stats.binom(trials, probability)

trials = [10, 20]
probability = [.7,.8,.9]

for x in trials:
    for y in probability:
        subjects = [(x / 2) - 1, (x - 2), (x - 1)]
        for z in subjects:
            print(round(stats.binom(x,y).sf(z),3))
```

```python
trials = [10, 20]
probability = [.7, .8, .9]

for x in trials:
    
    for y in probability:
        
        subjects = [[((x / 2) - 1),'sf','half stick around'], [(x - 2),'cdf','only 1 person leaves'], [(x - 1),'sf','all subjects stay']]
        
        for z in subjects:
            
            prob = y
            func = z[1]
            code = 'stats.binom(' + str(x) +',' + str(prob) + ').' + func + '(' + str(z[0]) + ')'
            print('Number of subjects:', x)
            print('Probability subject stays:', y)
            print('Scenario:',z[2])
            print('Function used:',code)
            print(round(eval(code),3))
            print('\n')
```
